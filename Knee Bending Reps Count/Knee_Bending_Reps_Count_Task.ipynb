```python
import numpy as np
import cv2
import math
import time
import PoseModule as pm
from datetime import datetime
```

```python
from datetime import datetime

cap = cv2.VideoCapture("KneeBendVideo.mp4")

# We need to set resolutions.
# so, convert them from float to integer.
frame_width = int(cap.get(3))
frame_height = int(cap.get(4))
   
size = (frame_width, frame_height)

result = cv2.VideoWriter('output_result.avi', 
                         cv2.VideoWriter_fourcc(*'MJPG'),
                         10, size)


detector = pm.PoseDetector()

qu = 0
duration = 8
count = 0
gd_count = 0
dir = 0
gd_dir = 0
bend_dir = 0

while True:
    
    ret, frame = cap.read()    
        
    start_time = datetime.now()
    diff = (datetime.now() - start_time).seconds # converting into seconds
    while( diff <= duration ):
        ret, frame = cap.read()
        
        
        frame = detector.findPose(frame,draw=False)
        #img = cv2.resize(img, (500,400))
        lmList = detector.findPosition(frame,draw=False)
        #print(lmList)
        if len(lmList) != 0:
            angle1 = detector.findAngle(frame,23,25,27)
            per = np.interp(angle1,(110,170),(0,100))
            
            gd_per = np.interp(angle1,(80,170),(0,100))

            #ber = np.interp(angle1, (180,140),(100,0))

            #print(angle1, per)

            if per == 0:
                if dir == 0:
                    count +=0.5
                    dir = 1

            if per == 100:
                if dir== 1:
                    count +=0.5
                    dir = 0

            #print(count)
            cv2.putText(frame, (f"Total Rep:{str(int(count))}"),(40,630),cv2.FONT_HERSHEY_PLAIN,5,(0,255,0),5)

            #cv2.putText(frame, str(int(count)),(45,650),cv2.FONT_HERSHEY_PLAIN,15,(0,255,0),15)
            
            if gd_per == 0 :
                if gd_dir == 0:
                    gd_count +=0.5
                    gd_dir = 1

            if gd_per == 100 :
                if gd_dir== 1:
                    gd_count += 0.5
                    gd_dir = 0
                    
            cv2.putText(frame,(f"perfect rep: {str(int(gd_count))}"),
                        (350,50),cv2.FONT_HERSHEY_PLAIN,3,(0,0,250),5)
            
            if angle1 < 140 and bend_dir == 0:
                start = datetime.now()
                bend_dir = 1
                
            if angle1 > 145 and bend_dir == 1:
                bend_dir = 0
                
                dur = (datetime.now()-start).seconds
                #print("dur",dur)
                
                #giving Feedback message
                if dur < duration:
                    cv2.putText(frame,(f"bend your knee for 8sec"),(270,230),
                                cv2.FONT_HERSHEY_SIMPLEX,1,(0,0,255),2,cv2.LINE_AA)
                    
                    print("Bend Your Knee for 8 Seconds")
                    
            cv2.putText(frame, (f"Total Knee Bending Time: {str(int(dur))}"), (270,160),
                        cv2.FONT_HERSHEY_SIMPLEX,1,(255,0,0),2, cv2.LINE_AA)
        

            

        #For showing the timer
        if angle1 < 130:
            if diff< duration:
                cv2.putText(frame, (f"Holding Timer:{str(diff)}"), (50,70), cv2.FONT_HERSHEY_SIMPLEX ,1,
                            (50,255,55), 2, cv2.LINE_AA)
                
        
        cv2.imshow('frame',frame)
        result.write(frame)
        diff = (datetime.now() - start_time).seconds

        k = cv2.waitKey(1)
        if k & 0xFF == ord("r"): # reset the timer
            break
        if k & 0xFF == ord("q"): # quit all
            qu = 1
            break
        
        
    if qu == 1:
        break
        
cap.release()
cv2.destroyAllWindows()
```
