# Import dependencies

```python
import numpy as np
import cv2
import math
import time
import PoseModule as pm
```

```python
cap = cv2.VideoCapture("video-1.mp4")

detector = pm.PoseDetector()

count = 0
dir = 0

while True:
    success, img = cap.read()
    #img = cv2.resize(img, (600,600))
    img = detector.findPose(img, draw = False)
    lmList = detector.getPosition(img, draw = False)
    
    if len(lmList) != 0:
        angle = detector.findAngle(img,12,14,16)
        per = np.interp(angle,(205,300),(0,100))
        bar = np.interp(angle,(210,300),(650,100))
        
        #print(angle, per)
        
        if per == 100:
            if dir == 0:
                count +=0.5
                dir = 1
                
        if per == 0:
            if dir== 1:
                count +=0.5
                dir = 0
                
        
        cv2.rectangle(img, (1100,200),(1175,300),(0,255,0),3)
        cv2.rectangle(img, (1100,int(bar)),(1175,300),(0,255,0),cv2.FILLED)
        cv2.putText(img, f"{int(per)}%",(45,670),cv2.FONT_HERSHEY_PLAIN,15,(0,255,0),25)
                
        #print(count)
        #cv2.rectangle(img, (0,450),(250,720),(0,255,0),cv2.FILLED)
        cv2.putText(img, str(int(count)),(45,650),cv2.FONT_HERSHEY_PLAIN,15,(0,255,0),15)
        
    
    
    cv2.imshow("Image", img)
    if cv2.waitKey(1) == 13:
        break
        
cap.release()
cv2.destroyAllWindows()
```
