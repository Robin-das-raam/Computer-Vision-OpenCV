## For individual RGB color


```python
img = cv2.imread('Trump.jpg')

B,G,R = cv2.split(img)
print(B.shape)

cv2.imshow("RED",R)
cv2.imshow("GREEN",G)
cv2.imshow("BLUE",B)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
#merging color channels

merged = cv2.merge([B,G,R])
cv2.imshow("merged",merged)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
## lets amplify the blue color

merged = cv2.merge([B+100,G,R])
cv2.imshow("merged",merged)
cv2.waitKey()
cv2.destroyAllWindows()
```

## cv.calcHist(images, channels, mask, histSize, ranges[, hist[, accumulate]])

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image1 = cv2.imread('E:/pictures/xy.jpg')
histogram = cv2.calcHist([image1],[0],None,[256],[0,256])

# ravel for flatten image array
plt.hist(image1.ravel(),256,[0,256])
plt.show();
```

```python
color = ('b','g','r')

for i,col in enumerate(color):
    hist2 = cv2.calcHist([image1],[i],None,[256],[0,256])
    plt.plot(hist2,color = col)
    plt.xlim([0,256])
plt.show();
```

```python
#application of mask

mask = np.zeros(image1.shape[:2],np.uint8)
mask[200:600,200:500] = 255
masked_img = cv2.bitwise_and(image1,image1,mask = mask)

#calculate histrogram with and without mask
hist_full = cv2.calcHist([image1],[0],None,[256],[0,256])
hist_mask = cv2.calcHist([image1],[0],mask,[256],[0,256])

plt.subplot(221)
plt.imshow(image1,'gray')

plt.subplot(222)
plt.imshow(mask,'gray')

plt.subplot(223)
plt.imshow(masked_img,'gray')

plt.subplot(224)
plt.plot(hist_full)
plt.plot(hist_mask)
plt.xlim([0,256])
plt.show()
```
