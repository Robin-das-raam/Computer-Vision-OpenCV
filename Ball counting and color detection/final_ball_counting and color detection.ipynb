# Color Detection & Ball Counting

```python
import cv2
import numpy as np
```

```python
def ball_count(frame):
    n = 0
    
    lower_all = np.array([0, 100, 100])
    upper_all = np.array([255, 255, 255])
    all_mask = cv2.inRange(hsv, lower_all, upper_all)
    
    lower_white = np.array([0, 0, 240])
    upper_white = np.array([255, 15, 255])
    w_mask = cv2.inRange(hsv, lower_white, upper_white)
    
    mask = all_mask+w_mask
    res = cv2.bitwise_and(frame_gau_blur,frame_gau_blur, mask = mask)
    ret,thresh = cv2.threshold(mask,127,255,3)
    contours,hirarchy = cv2.findContours(thresh,cv2.RETR_LIST,cv2.CHAIN_APPROX_SIMPLE)
    for contour in contours:
        area = cv2.contourArea(contour)
        if area >500:
            M = cv2.moments(contour)
            cx = int(M["m10"]/M["m00"])
            cy = int(M["m01"]/M["m00"])
            cv2.circle(frame,(cx,cy),3,(25,20,25),5)
            n +=1
    #print("num:",n)
            
    return n
```

```python
def get_col_ball(frame,mask):
    n = 0
    res = cv2.bitwise_and(frame_gau_blur,frame_gau_blur, mask = mask)
    ret,thresh = cv2.threshold(mask,127,255,3)
    contours,hirarchy = cv2.findContours(thresh,cv2.RETR_LIST,cv2.CHAIN_APPROX_SIMPLE)
    for contour in contours:
        area = cv2.contourArea(contour)
        if area >500:
            M = cv2.moments(contour)
            cx = int(M["m10"]/M["m00"])
            cy = int(M["m01"]/M["m00"])
            cv2.circle(frame,(cx,cy),3,(25,20,25),5)
            n +=1
            
    return n
```

```python
cap = cv2.VideoCapture("ball_sort2.mp4")


size = (700, 700)

result = cv2.VideoWriter('output_result2.mp4', 
                         cv2.VideoWriter_fourcc(*'MJPG'),
                         10,size)

while True:
    
    ret,frame = cap.read()
    frame = cv2.resize(frame,(700,700))
    
    hsv = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)
    frame_gau_blur = cv2.GaussianBlur(frame, (3, 3), 0)
    
    total_num_ball = ball_count(frame)
    print(str(total_num_ball))
    
    text = ("Total Number of Ball: " + str(total_num_ball))
    
    cv2.putText(frame, text, (20, 50), cv2.FONT_HERSHEY_SIMPLEX, 1, (100, 0, 255), 2)
    
    lower_white = np.array([0, 0, 240])
    upper_white = np.array([255, 15, 255])
    w_mask = cv2.inRange(hsv, lower_white, upper_white)
    white = get_col_ball(frame,w_mask)
    cv2.putText(frame, ("White Ball: " + str(white)), (20, 85), cv2.FONT_HERSHEY_PLAIN, 1.5, (0, 250, 0), 2)
    
    lower_green = np.array([45, 100, 50])
    upper_green = np.array([75, 255, 255])
    g_mask = cv2.inRange(hsv, lower_green, upper_green)
    green = get_col_ball(frame,g_mask)
    cv2.putText(frame, ("Green Ball: " + str(green)), (20, 105), cv2.FONT_HERSHEY_PLAIN, 1.5, (0, 255, 0), 2)
    
    lower_blue = np.array([85, 100, 100])
    upper_blue = np.array([126, 255, 255])
    b_mask = cv2.inRange(hsv, lower_blue, upper_blue)
    blue = get_col_ball(frame,b_mask)
    cv2.putText(frame, ("Blue Ball: " + str(blue)), (220, 80), cv2.FONT_HERSHEY_PLAIN, 1.5, (0, 0, 255), 2)
    
    lower_pink = np.array([0, 50, 70])
    upper_pink = np.array([10, 160, 250])
    p_mask = cv2.inRange(hsv, lower_pink, upper_pink)
    pink = get_col_ball(frame,p_mask)
    cv2.putText(frame, ("Pink Ball: " + str(pink)), (220, 100), cv2.FONT_HERSHEY_PLAIN, 1.5, (0, 0, 255), 2)
    
    lower_red = np.array([0,50,50])
    upper_red = np.array([10,255,255])
    mask0 = cv2.inRange(hsv, lower_red, upper_red)
    
    lower_red = np.array([170,50,50])
    upper_red = np.array([180,255,255])
    mask1 = cv2.inRange(hsv, lower_red, upper_red)
    
    red_mask = mask0+mask1
    
    all_red = get_col_ball(frame,red_mask)
    only_red = all_red - pink
    cv2.putText(frame, ("Red Ball: " + str(only_red)), (220, 120), cv2.FONT_HERSHEY_PLAIN, 1.5, (0, 0, 255), 2)
    
    lower_yellow = np.array([25, 50, 70])
    upper_yellow = np.array([35, 255, 255])
    y_mask = cv2.inRange(hsv, lower_yellow, upper_yellow)
    yellow = get_col_ball(frame,y_mask)
    cv2.putText(frame, ("Yellow Ball: " + str(yellow)), (460, 70), cv2.FONT_HERSHEY_PLAIN, 1.5, (0, 255, 0), 2)
    
    lower_purple = np.array([151, 155, 84])
    upper_purple = np.array([170, 255, 255])
    pur_mask = cv2.inRange(hsv, lower_purple, upper_purple)
    purple = get_col_ball(frame,pur_mask)
    cv2.putText(frame, ("Purple Ball: " + str(purple)), (460, 100), cv2.FONT_HERSHEY_PLAIN, 1.5, (0, 255, 0), 2)
    
    lower_orange = np.array([12,50,50])
    upper_orange = np.array([20,255,255])
    o_mask = cv2.inRange(hsv, lower_orange, upper_orange)
    orange = get_col_ball(frame,o_mask)
    cv2.putText(frame, ("Orange Ball: " + str(orange)), (460, 125), cv2.FONT_HERSHEY_PLAIN, 1.5, (0, 255, 0), 2)

    
    cv2.imshow("image",frame)
    result.write(frame)

    
    if cv2.waitKey(1) == 13:
        break
        
cap.release()
result.release()
cv2.destroyAllWindows()
    
    
```
