```python
import cv2

cap = cv2.VideoCapture(0)
cap.set(cv2.CAP_PROP_FRAME_WIDTH,1000)
cap.set(cv2.CAP_PROP_FRAME_HEIGHT,700)
#frame = cv2.imread("screen-0.webp")
while True:
    ret,frame = cap.read()
    h,w,_ = frame.shape
    hsv = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)
    #print(hsv)
    #frame = cv2.resize(frame,(700,700))
    #cv2.rectangle(frame,(250,600),(350,500),(255,255,255),2)
    cx,cy = int(w/2), int(h/2)
    cv2.rectangle(frame,(cx-30,cy-30),(cx+30,cy+30),(255,255,255),2)
    #cx,cy = int((250+350)/2), int((600+500)/2)
    #cv2.rectangle(frame,(250,450),(350,300),(255,0,0))
    
    
    pixel_cen = hsv[cy,cx]
    #print("pixel_cen",pixel_cen)
    hue_value = pixel_cen[0]
    
    color = "undefined"
    if hue_value < 5:
        color = "RED"
        
    elif hue_value < 22:
        color = "RED"
    
    elif hue_value < 33:
        color = "RED"
        
    elif hue_value < 78:
        color = "GREEN"
        
    elif hue_value < 131:
        color = "BLUE"
        
    else:
        color = "RED"
        
    bgr = frame[cy,cx]
    b, g, r = int(bgr[0]), int(bgr[1]), int(bgr[2])
    
    cv2.putText(frame,color, (10,70),0,1,(b,g,r),2)
    cv2.circle(frame,(cx,cy),4,(25,25,25),3)
    
    
    cv2.imshow("frame",frame)
    c = cv2.waitKey(1) 
    if c==27: 
        break
    
cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

frame = cv2.imread("unnamed.png")
gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
blur = cv2.GaussianBlur(gray, (11, 11), 0)
canny = cv2.Canny(blur, 30, 150, 3)
dilated = cv2.dilate(canny, (1, 1), iterations=0)

cnt, hierarchy = cv2.findContours(dilated.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
cv2.drawContours(rgb, cnt, -1, (0, 255, 0), 2)
print("coins in the image : ", len(cnt))
cv2.imshow("frame",frame)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import numpy as np
import cv2

img = cv2.imread('screen-0.webp')

gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
gray_blur = cv2.blur(gray, (3, 3))

#blurred = cv2.medianBlur(gray, 25) #cv2.bilateralFilter(gray,10,50,50)

minDist = 20
param1 = 50 #500
param2 = 30 #200 #smaller value-> more false circles
minRadius = 0
maxRadius = 0 #10

# docstring of HoughCircles: HoughCircles(image, method, dp, minDist[, circles[, param1[, param2[, minRadius[, maxRadius]]]]]) -> circles
circles = cv2.HoughCircles(gray_blur, cv2.HOUGH_GRADIENT, 1, 20, 50,30,2,40)

detect_cir = np.uint16(np.around(circles))

for (x,y,r) in detect_cir[0,:]:
    cv2.circle(img,(x,y),r,(0,0,255),3)

# if circles is not None:
#     circles = np.uint16(np.around(circles))
#     print(circles)
#     for i in circles[0,:]:
#         cv2.circle(img, (i[0], i[1]), i[2], (0, 255, 0), 6)

# Show result for testing:
cv2.imshow('img', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np
import cv2

image = cv2.imread('unnamed.png')
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

blur = cv2.medianBlur(gray, 5)

circles = cv2.HoughCircles(blur, cv2.HOUGH_GRADIENT, 1.5, 10,
                           param1=100,param2=100,minRadius=25,maxRadius=60)

circles = np.uint16(np.around(circles))

for i in circles[0,:]:
    # draw the outer circle
    cv2.circle(image,(i[0], i[1]), i[2], (255, 0, 0), 2)
    
    # draw the center of the circle
    cv2.circle(image, (i[0], i[1]), 2, (0, 255, 0), 5)

cv2.imshow('detected circles', image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
