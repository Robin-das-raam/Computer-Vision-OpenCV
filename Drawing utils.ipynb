### Task 1.1

**The task  is to get all the landmarks on the body from the image and save the image as result [image format]**

```python
import cv2
import mediapipe as mp
import time
```

```python
image = cv2.imread("unnamed.png")

cv2.imshow("input image",image)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img = cv2.imread("unnamed.png")

mpPose = mp.solutions.pose
pose = mpPose.Pose()

mpDraw = mp.solutions.drawing_utils

res = pose.process(img)

if res.pose_landmarks:
    mpDraw.draw_landmarks(img, res.pose_landmarks,mpPose.POSE_CONNECTIONS)


    for id, lm in enumerate(res.pose_landmarks.landmark):

        h, w, c = img.shape
        print(id , lm)

        cx,cy = int(lm.x*w), int(lm.y*h)
        cv2.circle(img, (cx,cy),4,(255,0,0),cv2.FILLED)
        

cv2.imshow("input", img)
cv2.imwrite("Output Image.jpg",img)
cv2.waitKey(0)
    
cv2.destroyAllWindows()
        
```
